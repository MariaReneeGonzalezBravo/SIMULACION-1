Tenemos el estimador dado por
$\pi = \frac{4n}{N}$
donde n es el número de dardos que caen dentro del semicirculo de radio 1 con centro en el origen y N es el número de lanzamientos.

La variable \( x_i \) se define como:

 \begin{cases}
1 & \text{si el dardo cae dentro del círculo}, \\
0 & \text{en cualquier otro caso}.
\end{cases}

donde la probabilidad de que caiga dentro del circulo es $P$ y en caso contrario es su complemento $1-P$

Ahora, consideramos:

$z = 4 \frac{x_1 + x_2 + ... + x_N}{N}$

Calculamos el valor esperado y la varianza:

$E[z]= \frac{4}{N} (E[x_1]+ ... + E[x_N])$
$E[z]= \frac{4}{N} (NP) = 4P$

Siendo $E[z]=4P$ un estimador ingresado (da el valor que esperabamos)

Calculando la varianza de $z$


$V[z]= \frac{16}{N^2} (V[x_1]+ ... + V[x_N])$
$V[z]= \frac{16}{N} (P(1 - P))$

Ahora podemos usar la desigualdad Chebyshev
$P(|z_N - \pi| >= \epsilon) <= \frac{V[z_N]}{ϵ^2}$

donde $\epsilon$ es considerado el error y desviación estandar

Siendo
$P(|z_N - \pi| >= \epsilon) <= \frac{16P(1-P)}{N \epsilon^2}$

luego, dado un valor de \epsilon podemos calcular el valor de $N$
