```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def volado():
        x = rnd.random()

        if 0 <= x <= 0.5:
            return "Sol"
        elif 0.5 <= x <= 1:
            return "Aguila"
```

```python
volado()
```

```python
def volados(n):
  prob=[]
  l=[]
  sol=0
  for i in range(n):
    if volado()=="Sol":
      sol +=1
      l.append(i)
      p= sol/(i+1) #probabilidad
      prob.append(p) # agragamos las probabiliades a la lista
  #grafica


  plt.plot(l,prob)
  plt.show()

  return prob
```

```python
y=volados(100)
np.mean(y) # media
#ley de los grandes numeros
```

```python
volados(10000000)
```
