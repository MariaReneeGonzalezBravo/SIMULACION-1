```python
def cdf(G):
  n=len(G)
  x=np.sort(G)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y =cdf(G)
```

```python
plt.step(X,Y) #GRAFICA DE DISTRIBUCIÓN
plt.axvline(media, color="green")
plt.axhline(0.5, color="red")
```
