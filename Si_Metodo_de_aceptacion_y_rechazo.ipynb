```python
#clase 07/04/24
#Usaremos teorema fundamental de Simulacion
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
import math as mt
#dadas las funciones

def f(x):
  return 20*x*(1-x)**3

def g(x):
  return 1

c=135/64


```

```python
def Y(a,b,n):
  listay=[]
  for i in range(n):
    Y= rnd.random()
    U= rnd.random()
    if U<= f(Y)/c:
      listay.append(Y)
  return listay
```

```python
def cdf(G):
  n=len(G)
  x=np.sort(G)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y =cdf(G)
```
